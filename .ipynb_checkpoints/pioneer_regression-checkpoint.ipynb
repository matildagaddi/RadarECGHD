```python
#Just for observing: linear regression 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate a synthetic electricity consumption dataset
np.random.seed(42)
date_rng = pd.date_range(start='1/1/2020', end='1/01/2021', freq='H')
data = pd.DataFrame(date_rng, columns=['timestamp'])
data['load'] = np.random.normal(500, 50, size=(len(date_rng)))  # Synthetic load values
data['temperature'] = np.random.normal(20, 5, size=(len(date_rng)))  # Synthetic temperature values

# Parameters
lags = 24  # Number of lags to consider

# Prepare the dataset with lagged features
def create_lagged_features(df, lags, target_col):
    for lag in range(1, lags + 1):
        df[f'lag_{lag}'] = df[target_col].shift(lag)
    df.dropna(inplace=True)
    return df

# Prepare data with lagged features
data = create_lagged_features(data, lags, 'load')

# Prepare features (X) and target (y)
X = data[['load', 'temperature'] + [f'lag_{i}' for i in range(1, lags + 1)]].values
y = data['load'].values

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict on the entire dataset
predictions = model.predict(X)

# Plot actual vs predicted values
plt.figure(figsize=(14, 7))
plt.plot(data['timestamp'], y, label='Actual Load')
plt.plot(data['timestamp'], predictions, label='Predicted Load', linestyle='--')
plt.title('Actual vs Predicted Load')
plt.xlabel('Timestamp')
plt.ylabel('Load (kW)')
plt.legend()
plt.tight_layout()
plt.show()

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, predictions)
print(f"Mean Squared Error: {mse}")
```

#HD regression model mainly inspired by RegHD paper (https://ieeexplore.ieee.org/document/9586284)

Assume you have a hypervector ( M ) of length ( D = 2k ). For regression using hypervectors:

$$ M = M + \alpha \cdot (y - M \cdot s) \cdot s $$

where:
- \( M \) is the hypervector being updated.
- \( $\alpha$ \) is a scalar parameter (learning rate).
- \( y \) is the actual value for regression.
- \( s \) is the hypervector corresponding to the prediction point.
- \( $\cdot$ \) denotes the dot product.

This formula updates \( M \) by adjusting it towards \( y \), scaled by the difference \( $y - M \cdot s$ \) and the direction \( s \).

```python
import numpy as np
import pandas as pd

# Generate synthetic electricity consumption dataset
np.random.seed(42)
date_rng = pd.date_range(start='1/1/2020', end='1/01/2021', freq='H')
data = pd.DataFrame(date_rng, columns=['timestamp'])
data['load'] = np.random.normal(500, 50, size=(len(date_rng)))  # Synthetic load values
data['temperature'] = np.random.normal(20, 5, size=(len(date_rng)))  # Synthetic temperature values

# Function to perform HDC regression update
def hdc_regression_update(M, s, y, alpha):
    update = alpha * (y - np.dot(M, s)) * s
    M = M + update
    return M

# Initialize hyperdimensional vectors M and s
D = 2000  # Dimensionality of the hypervectors
M = np.random.randn(D)  # Initialize M with random values
s = np.random.randn(D)  # Initialize s with random values

# Parameters
alpha = 0.01  # Learning rate
y = data['load'].iloc[0] 

# Perform HDC regression update
M = hdc_regression_update(M, s, y, alpha)

# Print updated M
print("Updated M vector:")
print(M)
```

```python
def train(model, epoch, optimizer):
    model.train()
    total_loss = 0
    correct = 0
    for batch_idx, (inputs, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = F.mse_loss(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        predicted = outputs.round()
        correct += (predicted == labels).sum().item()
    
    train_acc = correct / len(train_loader.dataset)

    return train_acc

def test(model):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            test_loss += F.mse_loss(outputs, labels).item()
            predicted = outputs.round() 
            correct += (predicted == labels).sum().item()
    
    test_acc = correct / len(test_loader.dataset)    
    return test_acc
```
